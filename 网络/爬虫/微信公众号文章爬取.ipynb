# 微信公众号文章爬取
- 注册一个微信公众号
- 新建一个群发
- 点击插入超链接
- 可以查询任意已知公众号对应的文章列表
- 根据获取接口，直接爬取

## 问题：
1. 点击下一页的等待时间设置为10秒，在爬取到700多条数据后，账号会被腾讯禁止查询，暂时不知道什么时候会解开

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.wait import WebDriverWait
import re
import random
import requests
import time
import math
import openpyxl
from datetime import datetime



def get_fakeid(name):
    '''
    获取公众号的id号
    :param query:
    :return:
    '''

    search_url = "https://mp.weixin.qq.com/cgi-bin/searchbiz?"

    query_id = {
        "action":"search_biz",
        "token":token,
        "lang":"zh_CN",
        "f":"json",
        "ajax":'1',
        "random":random.random(),
        "query":name,
        "begin":'0',
        "count":'5'
    }

    search_response = requests.get(search_url,cookies=cookie,headers=header,params=query_id)

    # 默认选择结果的第一个
    search_res = search_response.json().get('list')[0]

    # 公众号标识
    fakeid = search_res.get('fakeid')

    return fakeid


def get_max_page(fakeid):
    # 文章的获取
    doc_url = "https://mp.weixin.qq.com/cgi-bin/appmsg?"

    query_id_doc = {
        'token': token,
        'lang': 'zh_CN',
        'f': 'json',
        'ajax': '1',
        'random': random.random(),
        'action': 'list_ex',
        'begin': '0',
        'count': '5',
        'query': '',
        'fakeid': fakeid,
        'type': '9'
    }

    doc_request = requests.get(doc_url, cookies=cookie, headers=header, params=query_id_doc)

    # 最大文章数
    max_doc = doc_request.json().get('app_msg_cnt')

    # 循环次数
    num = math.ceil(int(max_doc) / 5)

    return num


# 我的公众号
user = ""
pwd = ""

# 要爬取的公众号列表
gzlist = ["妈妈帮","亲宝宝育儿","宝宝树"]

# 爬取的文章的开始时间,默认结束时间为今天
start_time = datetime.strptime('2018-01-01','%Y-%m-%d')

# 地址
url = "https://mp.weixin.qq.com/"

driver = webdriver.Chrome("./chromedriver.exe")

driver.get(url)

account_xpath = "//form[@class='login_form']//input[@name='account']"
account_el = WebDriverWait(driver,10).until(EC.visibility_of_element_located((By.XPATH,account_xpath)))
account_el.clear()
account_el.send_keys(user)

pwd_xpath = "//form[@class='login_form']//input[@name='password']"
WebDriverWait(driver,10).until(EC.visibility_of_element_located((By.XPATH,pwd_xpath))).send_keys(pwd)

login_xpath = "//form[@class='login_form']//a[text()='登录']"
WebDriverWait(driver,10).until(EC.visibility_of_element_located((By.XPATH,login_xpath))).click()

# 等待扫描二维码
time.sleep(5)
WebDriverWait(driver,30).until(EC.visibility_of_element_located((By.LINK_TEXT,"新建群发")))

# 重新刷新后才能获取到正确的cookie值，可以尝试刷新是否可以，本质上感觉和刷新一样
driver.get(url)
print("登录成功")

cookie = {}

# 获取cookie值
cookies = driver.get_cookies()
for i in cookies:
    cookie[i['name']] = i['value']

# 使用request重新登录
header = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/71.0.3578.98 Safari/537.36",
    "Host":"mp.weixin.qq.com"
}

response = requests.get(url,cookies=cookie)
token = re.findall(r'token=(\d+)',str(response.url))[0]


wb = openpyxl.load_workbook('./1.xlsx')

try:
    for name in gzlist:
        # 创建sheet
        ws1 = wb.create_sheet(name)
        ws1.append(['标题', '类别', '图片', '知识来源', '描述', '链接', '更新时间'])
        # 文章id
        gz_id = get_fakeid(name)
        # 总页数
        max_num = get_max_page(gz_id)
        count = 0
        for i in range(0,max_num+1):
            begin = i * 5 # 0,5,10,...
            # 文章的获取
            doc_url = "https://mp.weixin.qq.com/cgi-bin/appmsg?"

            query_id_doc = {
                'token': token,
                'lang': 'zh_CN',
                'f': 'json',
                'ajax': '1',
                'random': random.random(),
                'action': 'list_ex',
                'begin': str(begin),
                'count': '5',
                'query': '',
                'fakeid': gz_id,
                'type': '9'
            }

            doc_request = requests.get(doc_url, cookies=cookie, headers=header, params=query_id_doc)

            doc = doc_request.json()['app_msg_list']

            for item in doc:
                title = item['title']  # 题目
                item_type = ''  # 类别

                # 图片
                cover_name = '{}.png'.format(count)

                r = requests.get(item['cover'])
                with open("./img/{}".format(cover_name), 'wb') as f:
                    f.write(r.content)

                # 来源
                fr = '{}公众号'.format(name)

                # 描述
                desc = item['digest']

                # 链接
                link = item['link'].split("&chksm=")[0]

                # 更新时间
                update_time = datetime.fromtimestamp(item['update_time'])

                if update_time < start_time:
                    break

                count = count + 1

                ws1.append([title, item_type, cover_name, fr, desc, link, update_time])

            # 如果列表最后一个在开始时间之后，则跳出，不再找下一页
            last_time = datetime.fromtimestamp(doc[len(doc)-1]['update_time'])
            if last_time < start_time:
                break
            time.sleep(60)
except Exception as e:
    print(e)
    wb.save('./1.xlsx')

wb.save('./1.xlsx')
```
