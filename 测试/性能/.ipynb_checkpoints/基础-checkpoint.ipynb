# 性能测试
- 通过自动化的测试工具，模拟多种正常、峰值以及异常负载条件来对系统各项性能指标进行测试

## 分类
 - 负载测试
    - 定义：通过在被测系统上不断加压，直到性能指标达到极限
    - 目的：找到系统处理能力的极限

- 压力测试（强度测试）
    - 定义：测试系统在一定饱和状态下，系统能够处理的会话能力，以及系统是否会出现错误
    - 目的：检查系统处于压力性能下时，应用的表现

- 并发测试
    - 定义：通过模拟用户并发访问，测试多用户并发访问同一应用、模块或数据记录时是否存在死锁或其他性能问题
    - 目的：发现系统中可能隐藏的并发访问时的问题

- 配置测试
    - 定义：通过对被测系统的软硬件环境的调整，了解各种不同对系统的性能影响的程度，从而找到系统各项资源的最优分配原则
    - 目的：了解各种不同因素对系统性能影响的程度

- 可靠性测试
    - 定义：给系统加载一定业务压力的情况下，使系统运行一段时间，以此检测系统是否稳定
    - 目的：验证系统是否支持长期稳定的运行

## web性能指标
- 用户数量
    - 系统用户：该系统的注册用户数
    - 在线用户数：登录系统的用户
    - 并发用户数：系统可以同时承载的正常使用系统功能的用户数量，对服务器产生压力的数量
        - 并发：严格意义上，所有的用户在同一时刻做同一件事情或操作，这种操作一般做同一类型的业务
        - 并发：广义上，多个用户对被测系统发起了多个请求，这些请求可以是同一操作，也可以不同，属于混合场景
- 响应时间
    - 定义：事务可能由一系列请求组成，事务的响应时间主要是针对用户而言，属于宏观上的概念，是为了向用户说明业务响应时间而提出的
    - 公式：网络传输（请求）时间+服务器处理（响应）时间+前端展现时间
        - 3秒内：很快，5-10秒：还可以，10秒以上：无法忍受
- 吞吐量：单位时间内系统处理的请求数量，能直接反应服务器承受的压力
- 吞吐率：指用户在给定的一秒内从服务器获得的数据量
- TPS：每秒系统能抽离的交易或事务的数量，是衡量系统处理能力的重要指标
    - TPS大时，系统性能会比较好，但不可能无限大
- 资源利用率：对不同的系统资源的使用程度
    - 服务器CPU利用率：低于80%
    - 服务器内存利用率：低于85%
- 磁盘利用率
    - 作用：分析系统性能指标从而改善性能的主要依据
    - 关注：web服务器、数据库服务器、网络
- PV/UV/IP
    - PV：访问量，page view，页面浏览量或点击量，用户每次刷新即被计算一次（客户端\*页面数）
    - UV：独立访客，即Unique Visitor，访问您网站的一台电脑客户端为一个访客，00:00-24:00内相同的客户端只被计算一次（客户端数）
    - IP：独立IP，即Internet protocol，00:00-24:00内相同IP地址只被计算一次（IP数）
- 思考时间：3-5秒
