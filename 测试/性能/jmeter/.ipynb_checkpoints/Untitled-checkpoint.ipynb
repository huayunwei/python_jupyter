## Jmeter

* ### 特点

  * 多线程框架-支持请求并发操作

  * 用于对服务器、网络或对象模拟巨大负载

  * 支持数据库、FTP服务器等各类系统性能测试

  * 可以在多平台使用
* ### 下载和执行

  * 下载：[https://jmeter.apache.org/download\_jmeter.cgi地址](https://jmeter.apache.org/download_jmeter.cgi地址)

  ![](/assets/jmeter_download.png)

  * 执行

    * 解压

    * 执行apache-jmeter-4.0\bin\jmeter.bat

* ### 各元件介绍

  * 测试计划元件：用来描述一个性能测试，包含与本次性能测试所有相关功能

    * Threads（User）：线程，一般使用线程组

    * Test Fragment：测试片段，控制器上的一个特殊的线程组，在测试树上于线程组处于一个层级，与线程组不同的为它不被执行，除非被引用

  * 控制器：驱动处理一个测试

    * sample：取样器，性能测试中向服务器发送请求，记录响应信息，记录响应时间的最小单元

    * Logic Controller：逻辑控制器

      * sampler节点发送请求的逻辑顺序的控制器：if，switch，runtime controller，循环控制器

      * 组织可控制sampler节点：事务控制器，吞吐量控制器

  * 配置元件：提供对静态数据配置的支持

    * CSV Data Set Config：将本地数据文件形成数据池

  * 定时器：操作之间设置等待时间，思考时间

  * 前置处理器：用于在实际的请求发出之前对即将发出的请求进行特殊处理

  * 后置处理器：对Sampler发出请求后得到的服务器响应进行处理，一般用来提取响应中的特定数据

  * 断言：检查测试中得到的相应数据等是否符合预期

  * 监听器：监听元件，对测试结果数据进行处理和可视化展示

* ### 性能测试

  * 在计划下创建一个线程组

  * 在线程组下添加一个HTTP请求：

  * 在HTTP请求下建立查看结果树

  * 在HTTP请求下建立聚合报告

    * 参数介绍：单位为毫秒

      * samples：执行了几次请求

      * average：执行的平均相应时间，即响应信息中Load time的平均时间

      * median：一半的请求响应的平均时间

      * 90%Line：90%的响应的时间

      * min：请求最少花费的时间

      * max：请求最多花费的时间

      * error：请求错误率

      * throughput：吞吐量，服务器每秒请求数量

      * receive/send：每秒发送和接受的数据量

    ![](/assets/jmeter_聚合报告.png)

  * 仅一次控制器：不论线程组执行几次，该控制器下的请求只执行一次

    * 仅一次控制器在逻辑控制器中

  ![](/assets/jmeter_仅一次控制器.png)

  * switch controller：根据value随机执行下面的请求之一（每一个用户都会进行判断）

    * 通过switch controller中的value值来确定执行哪个请求
    * value给一个随机函数，根据value的值确定执行哪个请求
      * 如果value超过范围，则执行第0个
      * 如果value为空，则执行第0个
      * 如果value不为空，则根据value找对应的请求来执行
        * 如果没有找到任何相匹配的名字，则执行名字为default的请求
        * 如果没有default请求，则不执行任何请求

    ![](/assets/jmeter_switch_controller.png)

  * 循环控制器：循环控制器的下面在每次执行时，下面的内容循环执行指定次数

  * 参数设置

    * 读取文件的每一行，拆分他们作为参数，每个线程会读取不同的行

    * 配置元件-CSV数据文件设置

  ![](/assets/jmeter_参数设置.png)

  * 定时器：用于设置操作之间的等待时间，模拟用户的思考时间

    * 作用域：

      * 定时器在每个sampler之前执行，而不是只有，无论定时器在哪个位置

      * 如果仅作用于一个sampler则作为子节点加入

      * 如果需要执行后再等待，使用Test Action

    * 固定定时器：延迟多久之后执行下面的进程

    * 高斯随机定时器：延迟一定范围内的随机时间

    * 吞吐量定时器：Constant Throughput Timer

  * 监听器：用来对测试结果进行处理和可视化展示的一系列元件

  * 断言：用于检查测试中得到的相应数据等是否符合预期

  * 配置元件

    * HTTP请求默认值：线程组下的HTTP请求中参数为HTTP请求默认值中设置的参数

    * HTTP信息头管理器：添加消息头信息的，以key，value的形式

    * HTTP Cookie管理器：添加Cookie信息

    * HTTP Cache Manager：缓存管理器

  * 正则表达式处理器：用于获取请求响应数据中的值作为另外请求的参数

    * 后置处理其中-正则表达式处理器
