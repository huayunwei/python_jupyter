## 接口测试与postman

* 接口：泛指实体把自己提供给外界的一种抽象化物
* 分类：
  * 用户界面：人与程序之间的接口欧
  * 硬件接口：计算机各硬件组件自建的的接口
  * 软件接口：软件组件之间的接口
* 接口测试的原因
  * 接口是获取和操作资源的方式，大部分产品中，资源一般都是产品的核心
  * 接口中大部分内容是数据，通过数据的对比可以推测出产品的逻辑
* 测试流程
  * 需求评审（实际工作中一般没有）
  * 需求分析
  * 接口用例设计
  * 执行测试用例
  * BUG的定位于追踪
* 接口文档分析
  * 接口文档来源：开发
  * 内容
    * 接口名称
    * 接口地址
    * 支持格式（xml/json）
    * 请求方式
    * 请求参数（参数名称、类型、是否必填、参数说明等）
    * 返回参数（返回码，返回值信息，返回json消息或xml消息）
* app使用的协议依旧是http，https协议
* 用例设计
  * 正常场景：功能是否按照接口文档要求的功能实现
  * 异常场景
    * 参数为空
    * 参数多填
    * 参数少填
    * 参数类型错误
    * 参数数值范围错误
    * 异常场景下是否有正确的处理逻辑和响应提示
* 具体用例设计
  * 必填参数：对必填参数要设计传参数和不传参数，接口的返回情况
    * 必填参数都正确
    * 必填参数类型错误（非法，特殊字符）
    * 必填参数数值范围错误（超过边界）
    * 必填参数全部为空
    * 必填参数部分为空
  * 非必须参数：
    * 一般不会对非必须参数做非正确性的判断，所以只要测试接口返回的内容是否正确即可
  * 接口返回值
    * 比较返回码
    * 比较返回值的完整性，即返回的key全不全
    * 比较key的value数据类型
    * 比较key对应的value值
* 用例编写
  * ID
  * 接口名称
  * 接口连接
  * 测试名称
  * 返回格式
  * 请求方式
  * 参数
  * 参数设计
  * 预期结果
  * 返回数据类型
  * 实际结果
* 工具
  * [postman](/测试/开启测试之路/postman.md)
