## 功能测试

* UI测试  
  定义：用户界面测试：User interface Testing，指软件中的可见外观及其余用户交互的部分

  包括：菜单，对话框，窗口，其他控件

  目标：

  * 测试用户界面的风格是否满足客户要求，文字是否正确，页面是否美观，文字、图片组合是否完美，操作是否友好等
  * 确保用户界面通过测试对象的功能为用户提供相应访问或浏览功能。确保用户通过测试对象的功能为用户提供相应访问或浏览功能。确保用户界面符合公司或行业标准

  * 整体页面测试

    * 调整页面大小，检验页面的布局是否被压缩，重叠，覆盖等

    * 检查需要滚动条的页面是否提供滚动条，并可有效滑动

    * 页面被内容拉长，拉宽后，检查页面的文字是否还能正确显示，对齐方式，格式都保持一致，正常

    * 对外文和繁体的显示，需要支持，不允许出现乱码

    * 页面中tab翻页可以正常切换、支持tab按键切换的检验使用的正确

    * 页面推荐的内容要保证有结果，可以正常使用

  * 导航测试

    * 导航类型

      * 超链接

      * 菜单

      * 按钮

      * 对话框

      * 列表

      * 窗口

    * 对广告图片的点击是否正常，点击后给出的页面是否正确

    * 检查页面链接是否可正常点击，点击后链接到的位置是否正确

    * 链接点击后是否正常的发生颜色变化（若特殊处理也需要相应的检查）

    * 链接的打开方式是否合理（在当前窗口中打开，打开新窗口），是否符合产品设计

  * 超链接测试

    * 测试所有链接是否按指示的那样确实链接到该链接的页面

    * 测试所有链接的页面是否存在

    * 保证web应用系统没有孤立的页面，所谓孤立页面是指没有链接指向该页面

  * 图形测试

    * 要确保图形有明确的用途

    * web应用系统的图片尺寸要尽量小，并且要能清楚地说明某件事情，一般都链接到某个具体的页面

    * 验证所有页面字体的风格是否一致

    * 背景颜色应与字体颜色和前景颜色相搭配

    * 检查图片的尺寸、位置是否符合需求

    * 对广告图片的点击是否正常，点击后给出的页面是否正确

    * 页面上具有相同意义的图标应保持一致

    * 对链接其他网站的图片，无法显示时是否有容错性处理

  * 内容测试

    * 定义：检验web应用系统提供信息的正确性、准确性和相关性

  * 静态文字的测试

    * 检查字体、size、颜色、位置等因素是否符合需求
    * 页面文字图标的间距、行距是否统一，对齐方式是否统一
    * 含义是否符合要求
    * 是否存在拼写错误

  * 平台兼容性测试

    * windows
    * unix
    * mac
    * linux

  * 典型错误小结

    * 文字或图片重叠、剪裁、没有对齐
    * 重复菜单或链接
    * 提示信息错误、tooltip不显示
    * js错误问题
    * 页面链接错误
    * 页面访问资源不可用
    * 出现404错误
    * 按钮不工作、功能没有实现
    * 错误信息提示不合理
    * 文档无法下载

* 表单测试

* 什么是表单

web提供用户输入数据，通常采用输入框、按钮等形式，获取用户信息，并且提交到后台web服务器，web服务器会对收到的信息进行处理，这样的窗口叫做表单

* 表单测试点

  * 添加
    * 表单可以正常提交（数据格式正确）
    * 各个字段的含义清晰明了（根据需求）
    * 数据格式有合法性验证
    * 数据重复提交，如果一条已经成功提交，返回后再提交，看看系统是否做了处理
    * 验证数据正确性
    * 不能为空，对非空数据有明显标识或说明，对非空输入字段输入空数据有控制
      * 依次设置必填项为空，看是否有对应的提示信息
      * 输入全部必填项，验证是否可以正常提交
      * 输入全部需要填写的项，验证是否可以正常提交
  * 修改
    * 表单可以正常提交
    * 各个字段的含义清晰明了
    * 数据格式有合法性验证
    * 验证添加和修改对于相同字段要求的一致性
    * 重新验证修改正确性
    * 一条数据引用另外一条时，被引用的修改，引用数据是否一起改变
  * 查询
    * 非法范围的选取
    * 统计语句包含多个与或非条件
    * 条件逻辑紊乱
    * 分类查询、精确查询、模糊查询等能否完整列出满足条件的记录
    * 统计是当日、跨日、跨月、跨季、跨年度时，查询统计结果是否正确
    * 是否能按系统默认的条件查询
  * 删除
    * 可能造成严重后果的删除操作，是否有提示，是否可逆
    * 删除操作是否正确执行，检查文件或数据库数据
    * 删除多个或一个，检查是否可以正常执行
    * 是否可以删除一个正在使用的信息
    * 不选择任何记录，直接删除，是否会删除掉数据
    * 关闭删除的确认对话框或选择取消，是否会执行删除操作
    * 重新使用已删除的数据
  * 翻页
    * 首页、上一页、下一页、尾页
    * 总页数、当前页数
    * 指定跳转页
    * 指定每页显示的条数
  * 验证码
    * 目的：防止用户使用工具进行网站破坏性操作，验证用户合法身份
    * 输入正确的验证码能否正确通过
    * 重新进入验证码输入页面，检查验证码内容是否发生变化
    * 刷新页面，检查验证码是否有变化
    * 输入错误的验证码，是否有提示信息
    * 将大写的验证码改为小写
    * 验证码内容是否显示完整
    * 验证码显示速度
    * 能否手动更换验证码，更换速度
  * 表单元素
    * 文本框：数据的合法性
    * 密码框：长度、字符、加密显示
    * 日期：日期格式
    * 电话号码输入框
    * 邮件地址
    * 必填字段
    * 单选框
    * 复选框
    * 下拉列表：功能不同时，需要每个都测试
    * 表单按钮
    * 搜索、排序、分页
    * 共用页面session：多个tab页，一个退出，另外一个是否会继续能进行登录后的操作

* 超链接测试

  * 超链接

    * 定义：从某个网页元素指向一个目标的连接关系
      * 分类
        * 路径：内部连接和外部连接
        * 目标：
          * 文档：超链接的目标文件是文档，图片、视频、文本文件等
          * 锚点：单机超链接后可以跳转到本页面或其他页面指定位置
          * 电子邮件：打开浏览器默认的邮件处理程序进行发送电子邮件操作
          * 脚本：例如添加到收藏夹，设置为首页等
          * 空连接
    * 测试点

      * 超链接需要有明显的标识，鼠标放上去有形状变化、有下划线、鼠标放上去有提示，有动画提示等
      * 超链接可以正常打开，有效果，超链接打开的内容要正确
      * 有含有超链接对象的排版布局合理
      * 超链接点击前后颜色发生变化
      * 网站里面没有孤立页面
      * 打开超链接的响应时间合理
      * 超链接的对象的打开方式

    * 测试开始时间：在所有页面开发完成之后测试

    * 测试工具：xenu

      * 作用：测试连接目标是否存在和是否有孤立页面
      * 局限性：只能测试连接存不存在，无法确定连接跳转到正确的界面
