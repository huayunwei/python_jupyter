# 软件缺陷

## bug与缺陷
- bug是缺陷的一种表现形式，而一个缺陷是可以引起多种bug的

## 定义
* 软件未实现产品说明书要求的功能
* 软件出现了产品说明书指明不应该出现的错误
* 软件实现了产品说明书未提及到的功能
* 软件未实现产品说明书虽未提及但应该实现的功能
* 软件难以理解、不易使用、运行速度慢，或者软件测试员认为最终用户会认为不好

## 产生原因
* 产品说明书编写得不全面、不完善和不准确，而且经常更改，或整个开发组没有很好地沟通和理解
* 软件设计过程中片面、多变、理解与沟通不足
* 文档不足、进度压力、设计、编码错误都会引入软件缺陷
* 软件实施过程中安装、配置错误
* 并非所有软件缺陷都需要修复
* 市场的压力使得产品最终发行有时间限制
* 测试员错误理解或不正确操作引出的缺陷
* 错误的修改影响的模块较多，带来的风险较大
* 修改性价比太低，不值得修复
* 缺陷报告找那个提出的问题很难重现

##  如何发现软件缺陷
* 根据软件需求说明书发现软件缺陷
* 查找时间依赖和竞争条件的问题
* 查找边界条件、内存泄漏和数据泄漏
* 查找状态转换时出现的缺陷
* 查找资源依赖性：内存、网络、硬件等方面的缺陷
* 查找和硬件相关方面的缺陷

## 报告软件缺陷的原则
* 尽快报告软件缺陷
* 有效描述软件缺陷
* 在报告软件缺陷时不要做评价
* 对软件缺陷报告跟踪到底
* 每一个报告只针对一个软件缺陷

## 软件缺陷描述
* 缺陷ID
* 缺陷状态：新建、待解决、已解决、已验证
* 缺陷标题：描述缺陷的标题
* 缺陷详细描述：缺陷如何复现的步骤
* 缺陷的严重程度：致命、严重、一般、细微
* 缺陷的紧急程度：从1-4,1优先级最高，4最低
* 缺陷提交人
* 缺陷提交时间
* 缺陷所属项目、模块
* 缺陷指定解决人
* 缺陷指定解决时间
* 缺陷解决人
* 缺陷处理结果描述
* 缺陷复核人
* 缺陷复核结果描述
* 测试环境说明
* 必要的附件

## 缺陷严重性
* 严重性：软件缺陷对软件质量的破坏程度，即缺陷的存在对软件功能和性能产生这样的影响

* A：严重错误：不完全满足系统要求，基本功能未完全实现，或危机人生安全，系统崩溃或挂起等导致不能继续运行

    * 由程序引起的死机，非法退出

    * 死循环

    * 数据库发生死锁

    * 因错误操作导致的程序中断

    * 功能错误

    * 与数据库发生链接错误

    * 功能不符

    * 数据流错误

    * 严重的数值金额计算错误

* B：较严重错误：严重影响系统或基本功能的实现，且没有更正办法（重启软件不属于更正办法）系统不稳定、或破坏数据、或产生错误结果、或部分功能无法实现执行，而且是常规操作中不可避免的主要问题

    * 程序接口错误

    * 程序可被执行，但操作功能无法执行

    * 在小功能的某些项目（选项）使用无效（对系统非法致命的）

    * 功能实现不完整，如删除时没有考虑数据关联

    * 功能的实现不正确，如在系统实现的界面上，一些可接受输入的控件点击后无作用，对数据库的操作不能正确实现

    * 报表格式以及打印内容错误（行列不完整，数据显示不在所对应的行列等数据显示结果不正确的错误）

    * 轻微的数值计算错误

    * 界面错误（详细文档）

* C：一般性错误，严重影响要求后基本功能的实现，但存在合理更正办法（重新启动该软件不属于更正办法，系统或或响应时间变慢，产生错误的中间结果单不影响有限的问题）

    * 操作界面错误（包括数据窗口内列名定义、含义是否一致）

    * 打印内容、格式错误（只影响报表的格式或外观，不影响数据显示结果的错误）

    * 简单的输入限制未放在前台进行控制

    * 删除操作未给出提示

    * 虽然正确性不受影响，但系统性能和响应时间受到影响

    * 显示格式不正确

    * 增删改查功能，在本界面不能实现，但在另一个界面可补充实现

* D：建议优化，使操作者不方便或遇到麻烦，但它不影响执行工作功能或重要功能，界面拼写错误或用户使用不方便等小问题需要完善的问题

    * 界面不规范

    * 辅助说明描述不清楚

    * 输入输出不规范

    * 长时间操作未给用户提示

    * 提示窗口文字未采用行业术语

    * 可输入区域和只读区域没有明显的区分标志

    * 必填项与非必填项应加以区别

    * 滚动条无效

    * 键盘支持并不好，可在输入多行的字段中，不支持回车换行，或对相同字段，在不同界面支持不同快捷方式

    * 界面不能及时刷新，影响功能实现

    * 光标跳转设置不好，鼠标光标定位错误


## 优先级
* 最高优先级：软件的主要功能错误或造成软件崩溃，数据丢失的缺陷

* 较高优先级：影响软件功能和性能的一般缺陷

* 一般优先级：本地化软件的某些字符没有翻译或翻译不准确的缺陷

* 低优先级：对软件的质量影响非常轻微或出现几率很低的缺陷
