# web session界面：显示抓取的每一条http请求（每一条称为一个session）
![](img/fiddler_4.png)
- #：HTTP Request的顺序，从1开始，按照页面加载请求的顺序递增
![](img/fiddler_5.png)
- Result：HTTP响应状态
- Protocol：请求使用的协议
- Host：请求地址的域名
- URL：请求的服务器路径和文件名，包括get参数
- BODY：请求的大小，以byte为单位
- Caching：请求的缓存过期时间或缓存控制header等值
- Content-Type：请求响应的类型
- Process：发送此请求的windows进程即进程ID
- Comments：用户通过脚本或右键菜单给此session增加的备注
- Custom：用户可以通过脚本设置的自定义值

- Filter：多个维度的过滤条件
    - 部分1：Use Filters，勾选后进行过滤
    - 部分2：Actions
        - Run Filterset now：立即执行过滤条件
        - Load Filterset：加载保存的过滤设置
        - Save Filterset：保存过滤设置
        - help：帮助
    - 部分3：Hosts
        - 下拉框1：只显示内网或者外网选项
        - 下拉框2：根据主机名信息显示或隐藏或这标记指定请求
    - 部分4：Client Process
        - Show only traffic from ：根据进程信息进行过滤，选择后，将只显示由该进程发出的请求
        - show only Internet Explorer traffic：只关心由IE浏览器发出的请求
        - Hide traffic from service host：隐藏来自service host（由svchost.exe进程发出）的请求
    - 部分5：Request Headers
        - show only if URL contains：可以通过正则表达式过滤请求地址中包含或不包含的内容
        - Hide if URL contains：与上面的相似，只是隐藏
        - Flags requests with headers：支持指定某个http请求头名称，如果web session列表中存在该请求头，会加粗显示该session
        - Delete requests headers：与上面的相同，如果包含该请求头，则删除该请求头
        - Set request heade：将HTTP请求头更新为所输入的值
    - 部分6：Breakpoints断点设置
        - Break request on post：对post请求设置断点
        - Break request on get with query string：对所有GET且URL中包含了给定查询条件的请求设置断点
        - Break on XMLHttpRequest：对所有能够确定是通过XMLHTTPRequest的对象发送的请求设置断点
        - Break response on Content-Type：对所有响应头Content-Type中包含了指定文本的响应设置响应断点
    - 部分7：Response Status Code：根据响应状态码设置断点
    - 部分8：Response Type and Size
        - 根据响应数据的类型或隐藏
        - 根据响应数据的大小显示或隐藏
        - 根据响应所需要的时间设置背景颜色
        - 根据文件类型进行限制
    - 部分9：Response Headers 针对响应数据的头部进行过滤

# 命令行
![](img/fiddler_12.png)
- select：选择所有对应类型（content-type）为指定类型的HTTP请求如果选择图片：select image
- allbut：删除所有响应类型不是给定类型的HTTP请求
    - 如果要删除所有非image的session：allbut image
    - 等于keeponly image
- cls：清空全部
- ?text：查找URL，URL和问号后面的text值匹配则选中
- \>size或<size：响应大于size或小于size的
- =status：选择响应状态等于给定状态的所有HTTP请求
- @host：选择包含指定HOST的全部HTTP请求
- Bpafter xx：中断URL包含指定字符的全部session响应
- Bps xxx：中断HTTP响应状态为指定字符的全部session响应
- Bpv xxx：中断指定请求方式的全部session响应
- Bpm xxx：同Bpv
- Bpu xxx：同Bpafter
